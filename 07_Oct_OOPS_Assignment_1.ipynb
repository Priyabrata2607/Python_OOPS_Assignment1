Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
class vehicle_rental_system:
    def __init__(self):
        self.available_vehicles=[]

    def add_vehicles(self,vehicle):
        self.available_vehicles.append(vehicle)

    def rent_vehicle(self,customer_name,vehicle_type):
        for vehicle in self.available_vehicles:
            if vehicle.get_type() ==vehicle_type:
                self.available_vehicles.remove(vehicle)
                print(f'{customer_name} rented a {vehicle_type}')
                return True
        print(f'Sorry,no Available {vehicle_type} to rent.')
        return False

    def return_vehicle(self,vehicle):
        self.available_vehicles.append(vehicle)
        print(f'{vehicle.get_type()} returned')

    def display_available_vehicles(self):
        if not self.available_vehicles:
            print('No vehicles available.')

        else:
            print('Available vehicles:')
            for vehicle in self.available_vehicles:
                print(f'- {vehicle.get_type()}')


class vehicle:
    def __init__(self,vehicle_type):
        self.vehicle_type=vehicle_type

    def get_type(self):
        return self.vehicle_type



rent=vehicle_rental_system()

car1=vehicle('car')
bike1=vehicle('bike')
rent.add_vehicles(car1)
rent.add_vehicles(bike1)

rent.display_available_vehicles()

rent.rent_vehicle('Sudh','car')

rent.display_available_vehicles()

rent.return_vehicle(car1)

rent.display_available_vehicles()
```

Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
class Book:
    def __init__(self,title,author,isbn):
        self.title=title
        self.author=author
        self.isbn=isbn
        self.available=True

    def __str__(self):
        return f'{self.title} by {self.author} (ISBN : {self.isbn})'

class Library:
    def __init__(self):
        self.books=[]


    def add_book(self,book):

        self.books.append(book)

        print(f'Book {book.title} added to the library')


    def borrow_book(self,isbn):
        for book in self.books:
            if book.isbn==isbn and book.available:
                book.available=False
                print(f'Book {book.title} borrowed successfully.')
                return book

        print('Book not available for borrowing.')
        return None

    def available_book(self):
        available_books=[book for book in self.books if book.available]
        if not available_books:
            print('No available books in the library.')
        else:
            print('Available books are:')
            for book in available_books:
                print(f'- {book}')
            print()

library=Library()

book1=Book('Sherlock Homes','Arthur Conan Doyle','9781566196048')
book2=Book('Harry Pooter','J.K Rowling','978-0-7475-3269-9')
book3=Book('Lord of the rings','J.R.R Tolkien','9788845292613')

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

library.available_book()
library.borrow_book('9781566196048')

library.available_book()
```

Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
class product:
    def __init__(self,product_id,name,price,quantity):
        self.product_id=product_id
        self.name=name
        self.price=price
        self.quantity=quantity

    def __str__(self):
        return f"{self.name} (ID:{self.product_id},price:${self.price:.2f},quantity:{self.quantity})"

class Inventory:
    def __init__(self):
        self.products=[]

    def add_product(self,product):

        self.products.append(product)
        print(f'Product {product.name} added to the inventory.')

    def update_quantity(self,product_id,new_quantity):
        for product in self.products:
            if product.product_id==product_id:
                product.quantity=new_quantity
                print(f'Quantity of product "{product.name}" updated to {new_quantity}.')
                return
        print(f'Product with ID {product_id} not found in the inventory.')
        print()

    def display_available_product(self):

        if not self.products:
            print('No products in the inventory')

        else:
            print('Available products in the inventory:')
            for product in self.products:
                print(f'- {product}')
            print()

inventory=Inventory()

product1=product(1,'Laptop',50000,100)
product2=product(2,'Iphone',100000,1000)
product3=product(3,'speaker',10000,5000)

inventory.add_product(product1)
inventory.add_product(product2)
inventory.add_product(product3)

inventory.display_available_product()

inventory.update_quantity(1,50)
inventory.update_quantity(4,100)

inventory.display_available_product()

```

Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
class calculate_shape:
    def __init__(self,length,width,height=None):
        self.length=length
        self.width=width
        self.height=height


    def calculate_area(self):
        if self.height is None:
            return self.length*self.width
        else:
            return 2*(self.length*self.width+self.length*self.height+self.width*self.height)

    def calculate_perimeter(self):
        if self.height is None:
            return 2*(self.length+self.width)
        else:
            return 4*(self.length+self.width+self.height)


if __name__=='__main__':
    rectangle=calculate_shape(10,2)
    print("Area of rectangle:",rectangle.calculate_area())
    print('Perimeter of rectangle:',rectangle.calculate_perimeter())

    cuboid=calculate_shape(3,4,5)
    print('Area of cuboid:',cuboid.calculate_area())
    print('perimeter of cuboid:',cuboid.calculate_perimeter())
```

Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

```python
class studentmanagement:
    def __init__(self,studentID,name):
        self.student_id=studentID
        self.name=name
        self.grades=[]

    def add_grade(self,grade):
        self.grades.append(grade)

        print(f"Grade {grade} is added to the {self.name}'s record.")

    def calculate_average_grade(self):
        if not self.grades:
            print(f'No grades are available in {self.name}"s record.')
            return 0.0
        else:
            average=sum(self.grades)/len(self.grades)
            return average

    def display_student_details(self):
        print(f'StudentID: {self.student_id}')
        print(f'Name: {self.name}')

        if not self.grades:
            print(f'Grades not available in {self.name}"s record')
        else:
            print('Grades')
            for grade in self.grades:
                print(f'- {grade}')

            average_grades=self.calculate_average_grade()
            print(f'Average grades: {average_grades:.2f}')


student1=studentmanagement('192120','Priyabrata')
student2=studentmanagement('192107','Nitish Nayak')

student1.add_grade(80)
student1.add_grade(90)
student1.add_grade(95)

student2.add_grade(82)
student2.add_grade(88)
student2.add_grade(90)

student1.display_student_details()
print()
student2.display_student_details()
```

Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

```python
class socialmedia:
    def __init__(self,username):
        self.username=username
        self.posts=[]

    def add_posts(self,content):
        posts={'content':content,'author':self.username}

        self.posts.append(posts)
        print(f"Post added to the {self.username}'s Profile.")

    def display_posts(self):
        if not self.posts:
            print(f"No posts available in {self.username}'s Profile")
        else:
            print(f'posts from {self.username}')
            for post in self.posts:
                print(f"- {post['content']}")

    def search_posts(self,keyword):

        matching_posts=[post for post in self.posts if keyword.lower() in post['content'].lower()]

        if not matching_posts:
            print(f'No Posts by {self.username} contain the keyword {keyword}')
        else:
            print(f'Posts by {self.username} containing the keyword {keyword}')
            for post in matching_posts:
                print(f"- {post['content']}")

        return matching_posts

profile1=socialmedia('Priyabrata')
profile2=socialmedia('Kiran')

profile1.add_posts('I am Priyabrata')
profile1.add_posts('Python Developer')

profile2.add_posts('I am a Coder')
profile2.add_posts('IOS Developer')

profile1.display_posts()
profile2.display_posts()

profile1.search_posts('Priyabrata')
profile2.search_posts('hehe')
```

Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
from datetime import datetime

class todolist:
    def __init__(self):
        self.tasks=[]

    def add_task(self,task,due_date=None):

        new_task={"task":task,"due_date":due_date,"completed":False}

        self.tasks.append(new_task)

    def mark_tasks_completed(self,task_index):

        if 0<=task_index<=len(self.tasks):
            self.tasks[task_index]["completed"]=True
            print(f'Task "{self.tasks[task_index]["task"]}" marked as completed.')
        else:
            print('Invalid task index')


    def display_pending_tasks(self):
        pending_tasks=[task for task in self.tasks if not task['completed']]

        if not pending_tasks:
            print('No pending tasks')

        else:
            print('Pending tasks:')
            for index,task in enumerate(pending_tasks):
                print(f"{index+1}.{task['task']}")
                if task['due_date']:
                    print(f"  Due date: {task['due_date']}")
            print()


to_do_list=todolist()
to_do_list.add_task('Complete_coding','2023-12-10')
to_do_list.add_task('Buy grocories')
to_do_list.add_task('read a book')

to_do_list.display_pending_tasks()
to_do_list.mark_tasks_completed(1)
to_do_list.display_pending_tasks()
```
